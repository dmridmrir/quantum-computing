```python
from qiskit import *
from qiskit.visualization import *
from qiskit.providers.ibmq import *
from qiskit.providers.ibmq.job import job_monitor
from qiskit.providers.aer import AerSimulator
```

```python
# Z가 0일 경우 X값 출력, 1일 경우 Y값 출력 


X = QuantumRegister(1,"X")
Y = QuantumRegister(1,"Y")
Z = QuantumRegister(1,"Z")
Z_copy = QuantumRegister(1,"Z_copy")
out = QuantumRegister(1,"out")
res = ClassicalRegister(1,"res")

branch = QuantumCircuit(X,Y,Z,Z_copy,out,res)

branch.cx(2,3)
branch.x(2)
branch.ccx(1,3,4)
branch.ccx(0,2,4)

branch.measure(4,0)

branch.draw(output='mpl')
```

```python


X = QuantumRegister(1,"X")
Y = QuantumRegister(1,"Y")
Z = QuantumRegister(1,"Z")
Z_copy = QuantumRegister(1,"Z_copy")
out = QuantumRegister(1,"out")
res = ClassicalRegister(1,"res")

test = QuantumCircuit(X,Y,Z,Z_copy,out,res)

# X=0,Y=1,Z=1일 경우
test.x(1)
test.x(2)

test.cx(2,3)
test.x(2)
test.ccx(1,3,4)
test.ccx(0,2,4)

test.measure(4,0)

test.draw(output='mpl')
```

```python
sim = AerSimulator()

job = execute(test,sim,shots=1024)
result = job.result()
count = result.get_counts()
plot_histogram(count)
```
