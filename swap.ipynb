```python
from qiskit import *
from qiskit.visualization import *
from qiskit.providers.ibmq import *
from qiskit.providers.ibmq.job import job_monitor
from qiskit.providers.aer import AerSimulator
```

```python
qc = QuantumCircuit(2)

qc.cx(0,1)
qc.cx(1,0)
qc.cx(0,1)
qc.measure_all()
qc.draw(output='mpl')
```

```python
# q0 = 1, q1 = 0인 경우

test = QuantumCircuit(2)

test.x(0)

test.cx(0,1)
test.cx(1,0)
test.cx(0,1)

test.measure_all()
test.draw(output='mpl')
```

```python
sim = AerSimulator()

job = execute(test,sim,shot=10)
result = job.result()
counts = result.get_counts()
print(counts)
plot_histogram(counts)
```
